```python
import numpy as np
```

```python
#x는 공부시간, y는 성적
x=[2,4,6,8]
y=[81,93,91,97]
```

```python
#평균값
mx=np.mean(x)
my=np.mean(y)
print("x의 평균값: ",mx)
print("y의 평균값: ",my)

#기울기 공식의 분모
divisor = sum([mx-i**2 for i in x])
```

```python
#분자
def top(x,mx,y,my):
    d=0
    for i in range(len(x)):
        d+=(x[i]-mx)*(y[i]-my)
    return d

divided=top(x,mx,y,my)
print("분모: ", divisor)
print("분자: ", divided)
```

```python
# a, b 구하기
a=divided/divisor
b=my-(mx*a)
```

```python
print("기울기 a= ",a)
print("기울기 b= ",b)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
# 다중선형회귀 ( 추가 데이터 )

data=[[2,0,81],[4,4,93],[6,2,91],[8,3,97]]
x1=[i[0] for i in data]
x2=[i[1] for i in data]
y=[i[2] for i in data]
ax=plt.axes(projection='3d')
ax.set_xlabel('study_hours')
ax.set_zlabel('Score')
ax.dist=11
ax.scatter(x1,x2,y)
plt.show()
```

```python
x1_data=np.array(x1)
x2_data=np.array(x2)
y_data=np.array(y)

a1=0
a2=0
b=0
lr=0.02
epochs=2001
```

```python
# 경사하강법

for i in range(epochs):
    y_pred=a1*x1_data + a2*x2data +b #y
    error=y_data-y_pred #오차
    #오차함수 a 편미분
    a1_diff=-(2/len(x1_data)) * sum(x1_data * (error))
    a2_diff=-(2/len(x2_data)) * sum(x2_data * (error))
    #b 편미분
    b_diff = -(2/len(y_data)) * sum(error)
    
    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_diff
    
    if i % 500 == 0:
        print("epoch = %.f, 기울기1 = %.04f, 기울기2 = %.04f, 절편 = %.04f" % (i,a1,a2,b))
```
